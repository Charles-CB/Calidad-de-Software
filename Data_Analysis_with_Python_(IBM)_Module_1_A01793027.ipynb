Vamos a crearle encabezados a nuestro dataset, para ello vamos a crear una lista y despues se la vamos a asignar a las columnas

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

Asignamos nuestra lista a la columna

```python
df.columns = headers
df.head(10)
```

Remplazamos los signos "?" por valores nulos NaN de la librería de numpy

```python
df1 = df.replace("?",np.NaN)
```

Validamos que los valores  se hayan remplazado en nuestro nuevo dataset

```python
df1.head(10)
```

Ahora vamos a remover los valores nulos sobre la columna de precios, para ello vamos a utilizar dropna() y le vamos a poner sobre que columna queremos que nos elimine los datos nulos, para este caso será sobre la columna de price

```python
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```
